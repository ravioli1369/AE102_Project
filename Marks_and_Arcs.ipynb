## **Introduction**

India's largest engineering entrance exam, JEE (Joint Entrance Exam), has a lot of aspiring candidates trying their best at getting into the top engineering colleges in the country.The Joint Entrance Examination( also called JEE) is a national level entrance examination that is used to admit students into the top science and engineering colleges in the country. It is conducted in two levels- JEE Mains and JEE advanced. JEE Mains is the first stage which a student needs to clear in order to be eligible to appear for JEE advanced examination. It is conducted by the NTA(National Testing Agency). Apart from being a screening test for JEE advanced, the JEE Mains exam is used to admit students into the various NITs(National Institute of Technology) across the country.

After clearing JEE Mains, the top 2,50,000 candidates are selected to appear for the JEE advanced exam. It is mainly used as a basis for admission to the IITs(Indian Institute of Technology). But various other elite institutes such as IISc(Indian Institute of Science), IISERs(Indian Institute of Science Education and Research) and IIST(Indian Institute of Space Technology) also use the score obtained on the JEE-Advanced exam as the basis for admission. It is conducted by one of the seven zonal IITs(IIT Roorkee, IIT Kharagpur, IIT Delhi, IIT Kanpur, IIT Bombay, IIT Madras, and IIT Guwahati) under the guidance of the Joint Admission Board (JAB) on a round-robin rotation pattern.

The data that we used for our project is only for the JEE advanced exam.

Our aim is to try to understand the change in the way people think about engineering, their choice of department, and institute.
Here we look at the top 2000 kids of every year, and their choices; assuming that at these ranges of rank, a student will be entitled to a semblance of a choice for where they go. 

## Before heading on,

We will be referring to a numerical increase in rank as higher rank. Therefore, say, rank 2012 is a lower rank than 2021.

In all the graphs up ahead, there's an anomaly in year 2019: the number of seats are almost double to the other years. this is visible as a distinct spike in the graphs. This spike is due to data misrepresentation: while other years have only the general seats, 2019 has cumulative seats of all categories.

## Quantities of Interest and Scope

To organise data, we didn't take the number of seats as random variable since that isn't a representation of the students' choices but rather of the committee that fixes the number of seats. Our aim is to study how student mentality with respect to engineering career choices has progressed (or not) over the years.

The random variables of interest here are the year, the department and the institute the students went to. Since a student has more choice the higher their rank, we take a look at the top 6000 students from each year and their choice of department and institute. The assignment of RV is done via 'branch code' and 'institute code'. 

Our distribution would be a discrete random variable distributions for both quantities of interest. This limits us on a number of fronts (like calculation of point, interval estimates, linear regression) of data analysis. 

The hypothesis we would've wanted to go ahead with includes predicting the choice statistics of the future. Unfortunately this is out of the scope of this course. Hence, we will proceed with looking at the relations between our quantities of interest and draw analyses from the graphs of the same.

## Conclusion

Charting human decisions of the past to make future decisions more predictable has been going on from a very long time. Although we couldn't go far with the scope we had, we attempted at drawing some conclusions about the way people in our country perceive the branches and the institute. 

It is known that every institute has a good faculty for all courses. Yet, IIT Bombay has the highest number of interested candidates, drawing the top scorers there. If we sit down and look at the general population, IIT Bombay is the top choice only because it has always been the top choice in the past. The criteria people use for selecting an institute is the number of people who selected it in the past. A positive feedback loop seems to come into play. And hence we observe the gradual shifts in the **average ranks** of the most popular departments: CSE and Electrical.

Moving to departments, a similar reasoning is involved. CSE is the most sought after department because it was most sought before too. This is a bit disappointing that students don't try to consider what department might appeal to them by looking over the course content but rather they take the "most in-demand" department they can get into.
